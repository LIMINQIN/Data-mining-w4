# 离群点分析与异常检测

数据集：wave数据集  
来源： https://ir.library.oregonstate.edu/concern/datasets/47429f155?locale=en

```python
import numpy as np
import pandas as pd
import os
```

## 数据集中所有csv文件数据合并

```python
domain = os.path.abspath(r'/home/mqli/data-mining/wave/benchmarks/') #获取文件夹的路径
wave_pd = pd.DataFrame()
i=0
for info in os.listdir('/home/mqli/data-mining/wave/benchmarks/'):
    info = os.path.join(domain,info) #将路径与文件名结合起来就是每个文件的完整路径  
    data = pd.read_csv(info)
    print(i)
    wave_pd = wave_pd.append(data, ignore_index = True)
    i += 1
print(wave_pd)
```

```python
print(wave_pd)
wave_pd.to_csv("/home/mqli/data-mining/wave/benchmarks/all_file.csv",index=False)
```

## 缺失值处理，全部用0填充

```python
wave_pd = wave_pd.fillna(0)
wave_pd
```

## 处理标签数据（是否异常）

```python
wave_pd['ground.truth']=wave_pd['ground.truth'].map({'nominal':0,'anomaly':1})
wave_pd
```

## 使用pyod库中的PCA来训练异常检测模型

```python
from sklearn.model_selection import train_test_split
from pyod.models.pca import PCA
from pyod.utils.data import evaluate_print
from pyod.utils.example import visualize

x = wave_pd.drop(['point.id','motherset','origin','ground.truth'],axis=1)
y = np.array(wave_pd['ground.truth'])
random_seed = 5
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=random_seed)
clf_name = 'PCA'
clf = PCA()
clf.fit(x_train)

# get the prediction labels and outlier scores of the training data
y_train_pred = clf.labels_  # binary labels (0: inliers, 1: outliers)
y_train_scores = clf.decision_scores_  # raw outlier scores

# get the prediction on the test data
y_test_pred = clf.predict(x_test)  # outlier labels (0 or 1)
y_test_scores = clf.decision_function(x_test)  # outlier scores

# evaluate and print the results
print("\nOn Training Data:")
evaluate_print(clf_name, y_train, y_train_scores)
print("\nOn Test Data:")
evaluate_print(clf_name, y_test, y_test_scores)
```

从ROC指标来看，模型检测效果一般，有待提升，可能是选取的模型的原因，也有可能是特征选取的不够准确。于是尝试一下VAE模型来训练。

## VAE训练异常检测模型

```python
from pyod.models.vae import VAE

contamination = 0.1
clf_name = 'VAE'
clf = VAE(epochs=30, contamination=contamination, gamma=0.8, capacity=0.2)
clf.fit(x_train)

# get the prediction labels and outlier scores of the training data
y_train_pred = clf.labels_  # binary labels (0: inliers, 1: outliers)
y_train_scores = clf.decision_scores_  # raw outlier scores

# get the prediction on the test data
y_test_pred = clf.predict(x_test)  # outlier labels (0 or 1)
y_test_scores = clf.decision_function(x_test)  # outlier scores

# evaluate and print the results
print("\nOn Training Data:")
evaluate_print(clf_name, y_train, y_train_scores)
print("\nOn Test Data:")
evaluate_print(clf_name, y_test, y_test_scores)
```

```python
print("\nOn Training Data:")
evaluate_print(clf_name, y_train, y_train_scores)
print("\nOn Test Data:")
evaluate_print(clf_name, y_test, y_test_scores)
```
